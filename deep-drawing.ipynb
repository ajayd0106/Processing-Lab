```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('sample_1mm_wrinkled.csv')
data_2 = pd.read_csv('sample_3mm_optimized.csv')
```

```python
plt.plot(data_1['Stroke (mm)'],data_1['Force (kg)'])
plt.title('Force(kg) v/s Stroke(mm) for wrinkled sample',fontweight='bold')
plt.xlabel('Stroke(mm)')
plt.ylabel('Force(kg)')
plt.savefig('wrinkled.eps',format='eps')
```

```python
plt.plot(data_2['Stroke (mm)'],data_2['Force (kg)'])
plt.title('Force(kg) v/s Stroke(mm) for optimized sample',fontweight='bold')
plt.xlabel('Stroke(mm)')
plt.ylabel('Force(kg)')
plt.savefig('optimized.eps', format='eps')
```
