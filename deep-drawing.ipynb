```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
PATH1 = './datasets/sample_1mm_wrinkled.csv' # data obtained for wrinkled sample
PATH2 = './datasets/sample_3mm_optimized.csv' # data obtained for optimized sample
data_1 = pd.read_csv(PATH1)
data_2 = pd.read_csv(PATH2)
```

```python
plt.plot(data_1['Stroke (mm)'],data_1['Force (kg)'])
plt.title('Force(kg) v/s Stroke(mm) for wrinkled sample',fontweight='bold')
plt.xlabel('Stroke(mm)')
plt.ylabel('Force(kg)')
plt.savefig('./images/deep-drawing-wrinkled.png',format='png')
```

```python
plt.plot(data_2['Stroke (mm)'],data_2['Force (kg)'])
plt.title('Force(kg) v/s Stroke(mm) for optimized sample',fontweight='bold')
plt.xlabel('Stroke(mm)')
plt.ylabel('Force(kg)')
plt.savefig('./images/deep-drawing-optimized.png', format='png')
```
